# Univariate Analysis

Univariate analysis is a statistical method used to analyze a single variable in isolation. In this method, we examine and summarize the characteristics and properties of that specific variable without considering relationships with other variables. It's about understanding the distribution, frequency, and behavior of a single variable.

## Techniques used in Univariate Analysis:
Central Tendency Measures: Understanding the central or average value of the variable. This includes mean, median, and mode.

Dispersion Measures: Assessing the spread or variability of the variable. Measures like range, variance, standard deviation, and quartiles help understand the data's variability.

Frequency Distribution: Displaying how often each value in the variable occurs. Histograms, frequency tables, and density plots are used to represent this.

Summary Statistics: Providing a summary of the variable's characteristics using statistical methods, such as skewness, kurtosis, and percentiles.

Visualization Techniques: Plotting graphs or charts (like histograms, box plots, kernel density estimation plots) to visualize the distribution and characteristics of the variable.

Purpose of Univariate Analysis:
Data Understanding: It helps in understanding the basic properties and nature of the variable being analyzed.
Data Cleaning: Identifying anomalies or outliers within the variable.
Feature Selection: Selecting relevant variables for further analysis or modeling.
Initial Insights: Providing initial insights into the dataset before performing more complex analyses.
Univariate analysis forms the foundational step in exploratory data analysis, allowing for a comprehensive understanding of individual variables and their distributions.







```python
# Visualize the distribution of numerical variables
sns.histplot(data['Quantity'], bins=20, kde=True)
plt.title('Distribution of Quantity')
plt.show()

# Visualize categorical variables
sns.countplot(x='Country', data=data)
plt.title('Count of Orders by Country')
plt.xticks(rotation=90)
plt.show()
```

# Task 5: Analyze the sales trends over time. Identify the busiest months and days of the week in terms of sales.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Online Retail.csv')  

# Convert 'InvoiceDate' column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract month and day of the week from the 'InvoiceDate'
df['Month'] = df['InvoiceDate'].dt.month
df['Day_of_Week'] = df['InvoiceDate'].dt.day_name()

# Group by month and day of the week to get sales data
monthly_sales = df.groupby('Month')['Quantity'].sum()
weekly_sales = df.groupby('Day_of_Week')['Quantity'].sum()

# Plotting monthly sales trend
plt.figure(figsize=(10, 6))
monthly_sales.plot(marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Quantity')
plt.grid(True)
plt.xticks(range(1, 13))  # Assuming data spans a year
plt.show()

# Plotting sales by day of the week
plt.figure(figsize=(8, 5))
sns.barplot(x=weekly_sales.index, y=weekly_sales.values, order=[
            'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales Quantity')
plt.show()

# Identifying the busiest month
busiest_month = monthly_sales.idxmax()
print(f"The busiest month in terms of sales is {busiest_month}")

# Identifying the busiest day of the week
busiest_day = weekly_sales.idxmax()
print(f"The busiest day of the week in terms of sales is {busiest_day}")
```

# Task 6:  Explore the top-selling products and countries based on the quantity sold.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Online Retail.csv')  

# Group data by Product ID to find top-selling products
top_products = df.groupby('StockCode')['Quantity'].sum().sort_values(ascending=False).head(10)

# Group data by Country to find top-selling countries
top_countries = df.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plotting top-selling products
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top Selling Products')
plt.xlabel('Product ID')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.show()

# Plotting top-selling countries
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='salmon')
plt.title('Top Selling Countries')
plt.xlabel('Country')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.show()

# Displaying top products and countries
print("Top Selling Products:")
print(top_products)

print("\nTop Selling Countries:")
print(top_countries)
```
